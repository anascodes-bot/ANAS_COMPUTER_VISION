# Import Libraries

```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Original  Picture 

```python
image = cv2.imread(r'C:\Users\Dummy\Desktop\images (2).jpg')
cv2.imshow("Original Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Pixel Values To Pandas DataFrame

```python
image_df = pd.DataFrame(image.reshape(-1, 3), columns=['B', 'G', 'R'])
print('Basic Statistics of Image Pixel Values')
print(image_df.describe())
```

# Applying Binary Mask 

```python
height, width, _ = image.shape
```

```python
mask = np.zeros((height, width), dtype=np.uint8)
cv2.rectangle(mask, (100, 100), (400, 300), 255, -1)
bitwise_and = cv2.bitwise_and(image, image, mask=mask)
```

```python
cv2.imshow('Original Image', image)
cv2.imshow('Mask', mask)
cv2.imshow('Bitwise AND', bitwise_and)
           
cv2.waitKey(0)
cv2.destroyAllWindows()           
```

```python
fig = plt.figure(figsize=(10, 8))
rows = 2
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Original Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(mask, cmap='gray')
plt.axis('off')
plt.title("Mask")

fig.add_subplot(rows, columns, 3)
plt.imshow(cv2.cvtColor(bitwise_and, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Bitwise AND")
```
