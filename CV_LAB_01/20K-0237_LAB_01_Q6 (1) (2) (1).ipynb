# Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Original Picture

```python
image = cv2.imread(r'C:\Users\Dummy\Desktop\images.jpg')
cv2.imshow("Original Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Adding Text To Pictures

```python
target_width = 800
target_height = 600
resized_image1 = cv2.resize(image, (target_width, target_height))
text = "My Name is Anas"
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(resized_image1, text, (50, 150), font, 1.5, (0, 0, 255), 2)
```

# Black Image and Add Text to That

```python
image = np.zeros((300, 400, 3), dtype=np.uint8)

target_width = 800
target_height = 600

resized_image2 = cv2.resize(image, (target_width,target_height))

text = "My Name is Anas"
font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(resized_image2, text, (50, 150), font, 1.5, (0, 0,255), 2)
cv2.imshow('Image Containing Text', resized_image2)
           
cv2.waitKey(0)
cv2.destroyAllWindows()   
```

# SUBPLOT

```python
fig = plt.figure(figsize=(10, 8))
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(cv2.cvtColor(resized_image1, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Image with Text Overlay")

fig.add_subplot(rows, columns, 2)
plt.imshow(cv2.cvtColor(resized_image2, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Blank Image with Text Overlay")

plt.tight_layout()
plt.show()
```
