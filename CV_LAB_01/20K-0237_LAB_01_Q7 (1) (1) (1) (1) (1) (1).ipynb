# Import Libraries

```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Original Grey Scale Picture

```python
image = cv2.imread(r'C:\Users\Dummy\Desktop\istockphoto-1460209776-612x612.jpg')
cv2.imshow("Original Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# ImageThresholding

```python
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

cv2.imshow("Binary Thresholded Image", binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

image_df = pd.DataFrame(binary_image.reshape(-1), columns=['Pixel'])
print('Basic Statistics of Image Pixel Values')
print(image_df.describe())
```

# Rotation 60 Degree

```python
rotation_matrix =cv2.getRotationMatrix2D((image.shape[1] /2, image.shape[0] / 2), 45, 1)
rotated_image = cv2.warpAffine(image,rotation_matrix, (image.shape[1],image.shape[0]))
cv2.imshow('Rotated Image',rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# SUB PLOTS

```python
fig = plt.figure(figsize=(10, 8))
rows = 2
columns = 2


fig.add_subplot(rows, columns, 1)
plt.imshow(binary_image)
plt.axis('off')
plt.title("Gray Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(rotated_image)
plt.axis('off')
plt.title("Rotated Image")

plt.tight_layout()
plt.show()
```
