# Import Libraries

```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Original Picture 1

```python
image = cv2.imread(r'C:\Users\Dummy\Desktop\images (2).jpg')
cv2.imshow("Original Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Original  Picture 2

```python
image2 = cv2.imread(r'C:\Users\Dummy\Desktop\images (3).jpg')
cv2.imshow("Original Image", image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Image Blending

```python
image1 = cv2.resize(image, (image2.shape[1], image2.shape[0]))

alpha = 0.5
beta = 1 - alpha

blended_image = cv2.addWeighted(image1, alpha, image2, beta, 0)

cv2.imshow('Blended Image', blended_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Grey Scale Convert

```python
gray_image = cv2.cvtColor(blended_image,cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray Scale Image",gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# APPLY HISTOGRAM EQUALIZER

```python
equalized_image = cv2.equalizeHist(gray_image)
cv2.imshow('Equalized Image',equalized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig = plt.figure(figsize=(12, 8))
rows = 2
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Original Image 1")

fig.add_subplot(rows, columns, 2)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Original Image 2")


fig.add_subplot(rows, columns, 3)
plt.imshow(cv2.cvtColor(blended_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Blended Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(gray_image, cmap='gray')  
plt.axis('off')
plt.title("Gray Scale Image")

fig.add_subplot(rows, columns, 5)
plt.imshow(equalized_image, cmap='gray')  
plt.axis('off')
plt.title("Equalized Image")

plt.tight_layout()
plt.show()
```
