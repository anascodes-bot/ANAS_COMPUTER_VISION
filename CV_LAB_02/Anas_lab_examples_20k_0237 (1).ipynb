```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

**TASK 1 : Load CIFAR-10 Dataset and Display the Details of Datasets**

```python
#TASK 1
import tensorflow as tf
# Load CIFAR-10 dataset
(cifar_train_images, cifar_train_labels), (cifar_test_images, cifar_test_labels) = tf.keras.datasets.cifar10.load_data()
# Display dataset details
print('CIFAR-10 Train Images Shape', cifar_train_images.shape)
print("CIFAR-10 Train Labels Shape:", cifar_train_labels.shape)
print("CIFAR-10 Test Images Shape", cifar_test_images.shape)
print("CIFAR-10 Test Labels Shape:", cifar_test_labels.shape)
```

**TASK 2 : CIFAR-10 Visualization**

```python
# TASK 2
(cifar_train_images, cifar_train_labels), _ = tf.keras.datasets.cifar10.load_data()

plt.figure(figsize=(10, 3))
for i in range(5):
  plt.subplot(1, 5, i + 1)
  plt.imshow(cifar_train_images[i], cmap="gray")
  plt.title(f"Label: {cifar_train_labels[i]}")
  plt.axis("off")
plt.tight_layout()
plt.show()
```

**Task 4 : Prepare CIFAR-10 Dataset for Model training, testing and Validation, use 40,000 images for training, and
the remaining images are split between validation and testing**

```python
# TASK 3
(cifar_train_images, cifar_train_labels), (cifar_test_images, cifar_test_labels) = tf.keras.datasets.cifar10.load_data()
# Split dataset into training and testing
train_images, train_labels = cifar_train_images[:40000], cifar_train_labels[:40000]
val_images, val_labels = cifar_train_images[40000:], cifar_train_labels[40000:]
test_images, test_labels = cifar_test_images, cifar_test_labels
# Display the shapes of the split datasets
print("Train Images Shape:", train_images.shape)
print("Train Labels Shape:", train_labels.shape)
print("Validation Images Shape:", val_images.shape)
print("Validation Labels Shape:", val_labels.shape)
print("Test Images Shape:", test_images.shape)
print("Test Labels Shape:", test_labels.shape)
```

**TASK 4: Create Histograms and Bar Plots for CIFAR-10 Dataset**

```python
# TASK 4
import numpy as np
(_, cifar_train_labels), (_, cifar_test_labels) = tf.keras.datasets.cifar10.load_data()
# Calculate class frequencies
cifar_train_labels = cifar_train_labels.squeeze()
class_counts = np.bincount(cifar_train_labels)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(cifar_train_labels, bins=range(11), rwidth=0.8, align="left", density=True)
plt.xticks(range(10))
plt.xlabel("Class Label")
plt.ylabel("Frequency")
plt.title("CIFAR-10 Class Distribution Histogram")
plt.show()

# Plot bar plot
plt.figure(figsize=(8, 5))
plt.bar(range(10), class_counts, tick_label=range(10))
plt.xlabel("Class Label")
plt.ylabel("Frequency")
plt.title("CIFAR-10 Class Distribution Bar Plot")
plt.show()
```

**TASK 5 : Display Sample Images with Corresponding Labels for MNIST Dataset**

```python
# TASK 5


# Load CIFAR-10 dataset
(cifar_train_images, cifar_train_labels), _ = tf.keras.datasets.cifar10.load_data()

# Define class names for CIFAR-10
class_names = ["airplane", "automobile", "bird", "cat", "deer", "dog", "frog", "horse", "ship", "truck"]

# Display a few sample images with labels
plt.figure(figsize=(12, 6))
for i in range(10):
  plt.subplot(2, 5, i + 1)
  plt.imshow(cifar_train_images[i])
  plt.title(f"Label: {class_names[cifar_train_labels[i][0]]}")
  plt.axis("off")
plt.tight_layout()
plt.show()
```

**Task 6 : Perform Data preprocessing on the MNIST Dataset and Display the results.**

```python
(mnist_train_images, mnist_train_labels), (mnist_test_images, mnist_test_labels) = tf.keras.datasets.mnist.load_data()

# Data preprocessing for MNIST
mnist_train_images = mnist_train_images / 255.0
mnist_test_images = mnist_test_images / 255.0
mnist_train_labels = tf.keras.utils.to_categorical(mnist_train_labels, num_classes=10)
mnist_test_labels = tf.keras.utils.to_categorical(mnist_test_labels, num_classes=10)
# Display information about MNIST preprocessing
print("MNIST Train Images Shape:", mnist_train_images.shape)
print("MNIST Test Images Shape:", mnist_test_images.shape)
print("MNIST Train Labels Shape:", mnist_train_labels.shape)
print("MNIST Test Labels Shape:", mnist_test_labels.shape)
print("Sample MNIST Train Image:", mnist_train_images[0])
print("Corresponding MNIST Train Label:", mnist_train_labels[0])
```

**TASK 7 : Visualize MNIST data using bar plots (Summary Statistics)**

```python
import numpy as np
# Calculate summary statistics for MNIST
mnist_train_mean = np.mean(mnist_train_images)
mnist_train_std = np.std(mnist_train_images)
print("MNIST Train Data Summary Statistics:")
print("Mean:", mnist_train_mean)
print("Standard Deviation:", mnist_train_std)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Visualize MNIST data using bar plots
class_counts = np.bincount(mnist_train_labels)
class_names = [str(i) for i in range(10)]  # MNIST has 10 classes (0-9)
plt.bar(class_names, class_counts)
plt.xlabel("Class Names")
plt.ylabel("Frequency")
plt.title("MNIST Class Distribution")
plt.xticks(rotation=0)  # Remove rotation or set it to 0 for horizontal labels
plt.show()
```
